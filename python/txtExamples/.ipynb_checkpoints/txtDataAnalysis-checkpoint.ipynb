```python
# 필요한 패키지를 import
import pandas as pd
import glob # 경로와 이름을 지정하여 파일을 처리하는 모듈
import re # 메타 문자를 이용하여 특정 규칙을 작성하는 정규식 사용 모듈
from functools import reduce
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS, WordCloud
```

```python
# 데이터파일 병합
all_files = glob.glob('./data/myCabinetExcelData*.xls')
all_files
```

```python
all_files_data = [] # 저장할 리스트 생성
for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)
all_files_data[0]
```

```python
# all_files_data를 세로축을 기준으로 병합, all_files_data_concat 리스트에 저장
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index=True)
all_files_data_concat
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
all_files_data_concat.to_csv('./data/riss_bigdata.csv', encoding='utf-8', index=False)
```

```python
# 데이터 전처리
# 제목 컬럼 확인
all_title = all_files_data_concat['제목']
all_title
```

```python
# 전처리 진행: 
# 1. 영어가 아닌 단어는 제거하고 소문자로 정규화 ==> 단어 토큰화
# 2. 불용어를 제거하고 단어 형태 일반화를 위해 표제어 추출
stopWords = set(stopwords.words("english")) # 영어 불용어 목록 생성
lemma = WordNetLemmatizer() # 표제어 추출 객체 생성
words = []
for title in all_title: 
    EnWords = re.sub(r"[^a-zA-Z]+", " ",str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
print(words) # 전처리가 
```

```python
# 2차원 리스트를 1차원 리스트로 변환
words2 = list(reduce(lambda x, y: x+y, words))
print(words2)
```

```python
# 데이터 탐색, 분석 모델 구축
# 단어별 출현 횟수를 계산, 딕셔너리 객체 count를 생성
count = Counter(words2)
count
```

```python
# 출현횟수가 많은 상위 50개 단어 중에서 단어 길이가 1보다 큰 것만 저장
word_count = dict()
for tag, counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
        
```

```python
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = '85')
plt.show()
```

```python
all_files_data_concat['doc_count']=0
summary_year = all_files_data_concat.groupby('출판일', as_index = False)['doc_count'].count()
summary_year
```

```python
plt.figure(figsize=(12,5))
plt.xlabel("year")
plt.ylabel("doc-count")
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])
plt.show()
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='ivory', stopwords=stopwords, width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
